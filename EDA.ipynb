```python
def vlookup(reference_df, reference_colname, new_colname, matching_df, matching_colname, value_column_name):
    value_dict = {}
    
    get_data = matching_df[matching_colname]
    want_data = matching_df[value_column_name]
    for i in range(len(get_data)):
        if get_data[i] in value_dict.keys():
            pass
        else:
            value_dict[get_data[i]] = want_data[i]
    
    reference_df[new_colname] = reference_df[reference_colname].map(value_dict)
    
    return reference_df
```

```python
def time_between_orders(df, datecol, buyercol):
    sorted_df = df.sort_values([buyercol, datecol])
    dates = sorted_df[datecol].values
    buyers = sorted_df[buyercol].values
    
    times = [0]
    
    for i in range(1, len(dates)):
        if buyers[i] == buyers[i-1]:
            times.append(dates[i] - dates[i-1])
        else:
            times.append(0)
    
    sorted_df['times'] = times
    
    return sorted_df
```

```python
time_df = time_between_orders(simple_df, 'purchase-date', 'buyers')
```

```python
n_orders = time_df[time_df['times'] > 0]
```

```python
def time_between_first_second(df, buyercol, timecol):
    order_time_dict = {}
    order_times = df[timecol].values
    buyers = df[buyercol].values
    
    for i in range(len(order_times)):
        if buyers[i] in order_time_dict.keys():
            pass
        else:
            order_time_dict[buyers[i]] = order_times[i]
    
    return order_time_dict
```
